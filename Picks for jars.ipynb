```python
def table_graph(ax2, d, yr, title):
    a=[d[k:k+4] for k in range(0, len(d), 4)]
    df=pd.DataFrame()
    df[""]=["Q1","Q2","Q3","Q4"]
    t=0
    for j in a:
        df[t]=j
        t+=1
    year = [yr[k:k+4] for k in range(0, len(yr), 4)]
    l=["Quater"]
    for date in year:
        l.append(date[3].year)
    df.columns = l
    ccolors = np.full(len(df.columns), 'green')
    table = ax2.table(cellText=df.values,
                     cellLoc='center',
                     rowLoc='center',
                     colLabels=df.columns,
                     colColours=ccolors,
                     loc='center')
    table.scale(1,3.5)
    ax2.set_title(title, fontsize=22, y=0.79, pad=-2)
    ax2.axis('off')
    return ax1



def set_yearly_graph(ax4,dfd, endswithstr, company, title, setbottom=True,perc_to_hundred=1, 
                     splitby=1, ispickreport=False, needmarketshareline=False, 
                     titlefont=None, kind="line", color=None,ylim = None,startfromzero=False):
    if needmarketshareline:   # mktshare_2001_marketcap
        cols2 = ([col for col in dfd.columns if (col.endswith(endswithstr) \
                                                 and col.startswith("mktshare_") and ~('month' in col))])
    if title == 'Market cap':
        cols2 = ([col for col in dfd.columns if (col.endswith(endswithstr) \
                                                 and col.startswith("20") and ('month' not in col))])
    elif title == 'Profit':
        cols2 = ([col for col in dfd.columns if (col.endswith(endswithstr) \
                                                 and col.startswith("20") and ('ttm' not in col))])
    else:
        cols2 = ([col for col in dfd.columns if (col.endswith(endswithstr) \
                                                 and col.startswith("20"))])
    print("company is ", company)
    gp = dfd[dfd["Name"]==company][cols2].ffill(axis=1).bfill(axis=1)
    if len(gp)==0:
        gp = dfd[dfd["name"]==company][cols2].ffill(axis=1).bfill(axis=1)
    print("gp",gp)
    gpg = gp.transpose().dropna()
    print("gpg is =",gpg)
    if needmarketshareline:   # mktshare_2001_marketcap
        xvals = [int(ylabel.split("_")[1]) for ylabel in list(gpg.index)]
    else:
        xvals = [int(ylabel[:4]) for ylabel in list(gpg.index)]
    if 2023 in xvals:
        xvals.remove(2023)
        xvals = [2017] + xvals
        print('XVALS : ')
        print(xvals)
    gpg = perc_to_hundred * gpg
    if ispickreport:
        xvals = xvals[-6:]
        gpg = gpg.tail(6)    
    if kind=="bar":
#         print("gpg is =",gpg)
        col=list(gpg.columns)[0]
        yvals = list(gpg[col].ffill().values)
        print(col)
        print(title)
        print(yvals)
        print(xvals, yvals)
        if(ylim):
#             print('max value : ' + str(max(yvals)))
#             print(ylim)
            max_value = max(yvals)
            if(max_value <= 0.05 * ylim):
                ax4.set_ylim([0,0.15 * ylim])
            elif(max_value <= 0.1 * ylim):
                ax4.set_ylim([0,0.25 * ylim])
            elif(max_value <= 0.25 * ylim):
                ax4.set_ylim([0,0.5 * ylim])
            elif(max_value <= 0.4 * ylim):
                ax4.set_ylim([0,0.6 * ylim])
            else: 
                ax4.set_ylim([0,ylim])
        ax4.bar(xvals, yvals, color=color)
    else:
        if(startfromzero):
            max_value = list(gpg.max())[0]
            if(max_value < 5):
                extra_space = 1
            elif(max_value < 15):
                extra_space = 3
            elif(max_value < 40):
                extra_space = 5
            else:
                extra_space = 8
#             max_value = list(gpg.max())[0] 
            if(title == 'Return on Equity'):
                ax4.set_ylim([0,max_value + extra_space])
            elif(title == 'Return on Assets'):
                ax4.set_ylim([1,max_value + extra_space])
        if len(gpg)>1:
            print(title)
            print('XVALS :')
            
            print(xvals)
            print('GPG')
            print(list(gpg.values))
            ax4.plot(xvals, list(gpg.values))
    start, end = ax4.get_xlim()
    ax4.xaxis.set_ticks(np.arange(start, end+1, 2))
    ax4.xaxis.set_major_formatter(ticker.FormatStrFormatter('%d'))
    if titlefont:
        ax4.set_title(title, fontsize=titlefont,loc="left")
    else:
        ax4.set_title(title)
#     print(title, len(xvals), len(gpg))
#     print("done yearly", title)
    return ax4, xvals, gpg
#=================================================================================================================================

def set_quarterly_graph(ax1, dfd,xdates, startswithstr, company, title, do_reverse, 
        perc_to_hundred=1, subtract_one=0, setbottom=True, lastxyears=0, 
        ispickreport=False, linestyle="", titlefont=None, kind="line", color=None,label=None):
    plotarray = dfd.loc[(dfd["Name"]==company), dfd.columns.str.startswith(pat=startswithstr)].bfill(axis=1)
    if len(plotarray)==0:
        plotarray = dfd.loc[(dfd["name"]==company), dfd.columns.str.startswith(pat=startswithstr)].bfill(axis=1)

    plotarray = plotarray.sort_index(axis=1)
    if not xdates:
        LL=list(plotarray)
        xdates = get_dates_from_qs(LL)
    plotarray = plotarray.transpose()
    if do_reverse:
        plotarray = plotarray[::-1]
    plotarray = perc_to_hundred*(plotarray - subtract_one)
#     print(len(xdates), len(plotarray))
    if(len(xdates) < len(plotarray)):
        plotarray = plotarray[-len(xdates):]
    elif(len(xdates) > len(plotarray)):
        xdates = xdates[-len(plotarray):]
    if ispickreport:
        lastxyears = 5
#     yvals = [float(ylabel.split(startswithstr)[1]) for ylabel in list(plotarray.index)]
    col=plotarray.columns.tolist()[0]
    yvals = plotarray[col].tolist()
    years = mdates.YearLocator()
    ax1.xaxis.set_major_locator(years)
    if(title == 'Sales'):
        ylim_max = max(yvals[-lastxyears*4:])
        ylim_max = ylim_max * 1.3
        ylim_min = min(yvals[-lastxyears*4:])
        ylim_min = ylim_min * 0.6
        ax1.set_ylim([ylim_min,ylim_max])
        
    x = plotarray[-lastxyears*4:].values.tolist()
    if kind=="bar":
        t=ax1.bar(xdates[-lastxyears*4:], yvals[-lastxyears*4:], width=50, color=color,label=label)
    else:        
        t=ax1.plot(xdates[-lastxyears*4:],x, linestyle,label=label)
#     print(xdates)
#     if setbottom: 
#         ax1.spines['bottom'].set_position('zero') 
#     start, end = ax1.get_xlim()
#     ax1.xaxis.set_ticks(np.arange(start-1, end-1, 2))
#     ax1.xaxis.set_major_formatter(ticker.FormatStrFormatter('%d'))
    if titlefont: 
        ax1.set_title( title, fontsize=titlefont,loc="left") #company  + " - " +
    else:
        ax1.set_title(title)
#     print(title, len(xdates[-lastxyears*4:]), len(plotarray[-lastxyears*4:]))
#     print("done qtrly", title)
    return t[0],ax1, xdates[-lastxyears*4:], yvals[-lastxyears*4:],title,titlefont,color,label



# COSMOFIRST

# SHARDAMOTR
# DHANUKA

notapplicable=0
jarspicks="""FCL
RHIM
ITC
BHEL""".split('\n')
i=0

with PdfPages('testing_data_89q.pdf') as pdf_page_picks:
    main_img = mpimg.imread("jama_wealth_logo@3x.png")
    fig, axs = plt.subplots(nrows=2,ncols=3, figsize=(22,22), #constrained_layout=True)
                            gridspec_kw={'hspace': 0.3, 'wspace': 0.3})
#     figs, axss = plt.subplots(nrows=2,ncols=2, figsize=(22,22), #constrained_layout=True)
#                             gridspec_kw={'hspace': 0.3, 'wspace': 0.3})
    
    plt.rcParams.update({'font.size': 22})
#     figs = plt.figure(figsize=(22, 22))
    for nsec in jarspicks:
        i+=1
        company_L = dfd.loc[(dfd["SYMBOL"]==nsec)]["Name"].tolist()
        if np.nan in company_L:
            company_L = dfd.loc[(dfd["SYMBOL"]==nsec)]["name"].tolist()
        if company_L:
            company = company_L[0]
        else:
            print(nsec, ": Not found")
            notapplicable += 1
            continue
        print("==========================")
        print("company ===", company)
        sector = dfd.loc[(dfd["SYMBOL"]==nsec)]["sector"].tolist()[0]
        print(sector)
        print("-*"*40)
        is_bfsi = True if sector in ["Bank","Banks","Finance"] else False
        # Code for right corner logo
        logo = fig.add_axes([0.825, 0.9, 0.1, 0.1], anchor='NE')    
        logo.imshow(main_img)
        logo.axis('off')
        # In this method, adjust 1.2 or 1.47 for text spacing which is dependent on text length
        # and adds space vertically. Removing facecolor changes textbox to blue so had to 
        # explicitly define white to make background transparent.
#         fig.text(0,1.47, "Company ABC", fontsize=26)
        
        cotitle = "%d. %s [%s]" % (i, company, sector)
        print("Pick:", cotitle)
        fig.suptitle(cotitle, fontsize=27, fontweight='bold', y=0.97)
        # Method also creates a black outline for text. 
        # Add \n in string whenever you require line breaks and also to adjust page becoming smaller because of lengthy text.
        # 113 lines of text.
        txt = "Jama exists to make people's live better by helping them grow their wealth. And keep more of that growth with themselves, by avoiding commissions & hidden fees."
        """\ We are doing this with a passion to bring a revolution in the  personal finance 'industry'. We are pained that many customers get a raw deal from \
        brokers, banks and other advisors. Commissions worth thousands of crores are paid,\
        but do customers get every penny worth of it? We are breaking the old model with \
        technology that changes the investment process. Makes it simple. No commissions, \
        no high-fees, no conflict of interest. Jama is about simplifying money and growing it \
        for you. Our model is simple."""
        fig.text(0.03, 0.85, txt ,fontsize=18)
        ispickreport = True
        ax1 = axs[0,0]
        ax1.clear()
        net_worth_col = list(dfd.columns[dfd.columns.str.contains('_networth')])
    
        try:
            company_net_worth = list(dfd[dfd['Name'] == company][net_worth_col].ffill(axis=1).bfill(axis=1).values)[-1].max()
        except:
            print(nsec)
            print(list(dfd[dfd['SYMBOL'] == nsec][net_worth_col].ffill().values))
            company_net_worth = list(dfd[dfd['SYMBOL'] == nsec][net_worth_col].ffill(axis=1).bfill(axis=1).values)[-1].max()
        print('ylim')
        print("comp networth is here==",company_net_worth)
        
        #Networth
        set_yearly_graph(ax1,dfd, endswithstr="_networth", company=company, title="Net Worth", 
                         ispickreport=ispickreport, titlefont=30, kind="bar", color="darkblue")
        
        #Debt or ROA
        ax2 = axs[0,1]
        ax2.clear()
        if is_bfsi:
            set_yearly_graph(ax2,dfd, endswithstr="_roa", company=company, title="Return on Assets", 
                             perc_to_hundred=100, ispickreport=ispickreport, titlefont=30,
                             color="lightred",startfromzero=True)
        else:
            set_yearly_graph(ax2, dfd,endswithstr="_debt", company=company, title="Debt", 
                             ispickreport=ispickreport, titlefont=30, kind="bar", color="red",ylim = company_net_worth)
            
        #SALES    
        ax3 = axs[0,2]
        ax3.clear()
        if is_bfsi:
            sales = set_quarterly_graph(ax3,dfd,xdates,"totalincome_q", company, setbottom=True, title="Total Income", 
                        do_reverse=False, ispickreport=ispickreport, titlefont=30, kind="bar", color="lightblue")
        else:
            sales = set_quarterly_graph(ax3,dfd,xdates,"sales_q", company, setbottom=True, title="Sales",
                        do_reverse=False, ispickreport=ispickreport, titlefont=30, kind="bar", color="lightblue")
            
        #Market Cap
        ax4 = axs[1,0]
        ax4.clear()
        set_yearly_graph(ax4,dfd, endswithstr="_marketcap", company=company, title="Market cap", 
                         ispickreport=ispickreport, titlefont=30, kind="bar", color="pink")
        
        #PROFIT
        ax6 = axs[1,2]
        ax6.clear()
        if is_bfsi:
            profit = set_quarterly_graph(ax6,dfd, xdates, "netprofit_q", company, setbottom=True, title="Profit", 
                    do_reverse=False, ispickreport=ispickreport, titlefont=30,kind="bar", color="lightgreen")
        else:
            profit = set_quarterly_graph(ax6, dfd,xdates, "pat_q", company,setbottom=True,  title="Profit", 
                    do_reverse=False, ispickreport=ispickreport, titlefont=30,kind="bar", color="lightgreen")
        
#         #ROE
#         ax5 = axs[1,1]
#         ax5.clear()
#         if is_bfsi:
#             set_yearly_graph(ax5,dfd, endswithstr="_roe", company=company, title="Return on Equity", 
#                              perc_to_hundred=100,
#                              ispickreport=ispickreport, titlefont=30, color="skyblue",startfromzero=True)
#         else:
#             set_yearly_graph(ax5, dfd,endswithstr="_ronw", company=company, title="Return on Equity", 
#                              ispickreport=ispickreport, titlefont=30, color="skyblue",startfromzero=True)
            
        #ROE_Qt
        ax5 = axs[1,1]
        ax5.clear()
        if is_bfsi:
            set_quarterly_graph(ax5,dfd, xdates, "roe_q", company, setbottom=True, title="Return on Equity", 
                    do_reverse=False, ispickreport=ispickreport, titlefont=30,kind="line", color="lightblue")
        else:
            set_quarterly_graph(ax5, dfd,xdates, "roe_q", company,setbottom=True,  title="Return on Equity", 
                    do_reverse=False, ispickreport=ispickreport, titlefont=30,kind="line", color="lightblue")
#          
        
        fig.subplots_adjust(top = 0.77)
        plt.show()
        
        # Method to label x-axis and y-axis. Change as necessary.
        for ax in axs.flat:
            if ax == axs[1,2]:
                ax.set(xlabel='') #, ylabel='y-label')
            elif ax == axs[1,0] or ax == axs[1,1]:
                ax.set(xlabel='') #, ylabel='y-label')
        #     elif ax == axs[0,0] or ax == axs[0,1] or ax == axs[0,2]:
        #         ax.set(ylabel='y-label')
        #  Method to hide x & y label to show graphs sharing x & y axis in vertical orientation.   
        """for ax in axs.flat:
            if ax == axs[0,0] or ax == axs[0,1] or ax == axs[0,2]:
                plt.setp(ax.get_xticklabels(), visible=False)"""
        pdf_page_picks.savefig(fig, papertype='a4', orientation='landscape')
        
        # =================================================================================
        
#         # table data creating starting 
#         print("company is ", cotitle)
#         figs, ax= plt.subplots(nrows=2,ncols=1, figsize=(22,22), gridspec_kw={'hspace': 0.3, 'wspace': 0.3})
#         plt.rcParams.update({'font.size': 22})
#         logo = figs.add_axes([0.825, 0.9, 0.1, 0.1], anchor='NE')
#         logo.imshow(main_img)
#         logo.axis('off')
#         figs.suptitle(cotitle, fontsize=27, fontweight='bold', y=0.97)
#         figs.text(0.03, 0.85, txt ,fontsize=18)
        
#         #Sales table
#         if sales[3]:
#             ax1 = ax[0]
#             ax1.clear()
#             table_graph(ax1, sales[3],sales[2], "SALES REPORT")
            
#         #Profit table
#         if profit[3]:
#             ax2 = ax[1]
#             ax2.clear()
#             table_graph(ax2, profit[3],profit[2], "Profit REPORT")
            
#         fig.subplots_adjust(top = 0.77)
#         plt.show()
#         pdf_page_picks.savefig(figs, papertype='a4', orientation='landscape')
        
       
```
